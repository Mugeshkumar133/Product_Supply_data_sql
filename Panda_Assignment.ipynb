```python
import pandas as pd
purchase_df=pd.read_csv("/content/Assignment-1_Data (1).csv",sep=";")
purchase_df
```

```python
purchase_df.info() #For initial checking
```

```python
purchase_df.dropna(subset=["Itemname"])
purchase_df.dropna(subset=["Country"])
purchase_df.dropna(subset=["CustomerID"])
```

```python
purchase_df.columns
```

```python
# Let we convert the price in str format to float type for calculation purposes as we have the quantity in float type
```

```python
def convert(x):
  a=str(x)
  b=a.replace(",",".")
  c=float(b)
  return c
purchase_df["Price"]= purchase_df["Price"].apply(lambda x: convert(x))

```

```python
purchase_df["Price"]
```

```python
purchase_df.info()
```

```python
# Q1: Which product was frequently bought?
```

```python
frequently_sold_items=purchase_df.loc[purchase_df.groupby("Itemname")["Quantity"].transform(sum).idxmax()]
frequently_sold_items
```

```python
#Answer1: Frequently sold items is   "MEDIUM CERAMIC TOP STORAGE JAR"      | Total Quantity sold is 74215.0
```

```python
purchase_df.dropna(subset=["Quantity"],inplace=True)
purchase_df.dropna(subset=["Price"],inplace=True)
```

```python
purchase_df["Total Price"]=purchase_df["Price"]*purchase_df["Quantity"]
purchase_df["Total Price"]
```

```python
# How much on an average does the customer spent ?
```

```python
average_spent=purchase_df.groupby("CustomerID")["Total Price"].mean().reset_index()[["CustomerID","Total Price"]]
average_spent.sort_values(by="Total Price",ascending =False)
```
